```python
# 1. Initialize connection
client = MongoClient('mongodb://localhost:27017/')
db = client['UniMediTrend']
collection = db['clinic_logs']

# 2. Idempotency Check: Clear the collection to prevent duplicates
collection.delete_many({})
print("Cleared existing records in 'clinic_logs'...")

# 3. Convert the pandas DataFrame to BSON format
records = df.to_dict('records')

# 4. THE FIX: Assign the insert operation to a variable named 'result'
# This traps the massive list of ObjectIds inside the variable so they don't print!
result = collection.insert_many(records)

# 5. Verify the insertion cleanly
print(f"Success! {collection.count_documents({})} records securely loaded into the 'UniMediTrend' database.")
```
